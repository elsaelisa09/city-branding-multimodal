```python
from selenium import webdriver
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
import time
import csv
import wget
import os
```

```python
# open the webpage
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

driver.get("https://www.instagram.com")

wait = WebDriverWait(driver, 15)

# target username
username = wait.until(EC.presence_of_element_located((By.NAME, "email")))
password = wait.until(EC.presence_of_element_located((By.NAME, "pass")))
# enter username and password
username.clear()
username.send_keys("") #MASUKIN USERNAME IG
password.clear()
password.send_keys("") #  MASUKKIN PASSWORD IG

login_btn = wait.until(EC.element_to_be_clickable(
    (By.XPATH, "//div[@aria-label='Log In']")
))
login_btn.click()

time.sleep(5)
# We are logged in!
```

```python
import time

hastag = "wisatalampungselatan"
driver.get("https://www.instagram.com/explore/search/keyword/?q=%23" + hastag )

time.sleep(3)
```

```python
n_scrolls = 5000
post_links = set()

for _ in range(n_scrolls):
    anchors = driver.find_elements(By.TAG_NAME, 'a')
    for a in anchors:
        href = a.get_attribute('href')
        if href and "/p/" in href:
            post_links.add(href)

    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(3)

print("Total post:", len(post_links))
```

## Percobaan 1

```python
data = []
folder = "DatasetFoto2"

for idx, link in enumerate(post_links):
    driver.get(link)
    time.sleep(5)

    # --- CAPTION ---
    try:
        caption = driver.find_element(By.XPATH, "//article//h1").text
    except:
        caption = ""

    # --- TANGGAL ---
    try:
        date = driver.find_element(By.TAG_NAME, "time").get_attribute("datetime")
    except:
        date = ""

    # --- LIKE ---
    try:
        like = driver.find_element(By.XPATH, "//section//span").text
    except:
        like = ""

    # --- LOKASI ---
    try:
        location = driver.find_element(By.XPATH, "//a[contains(@href,'/explore/locations/')]").text
    except:
        location = ""

    # --- FOTO (lebih spesifik) ---
    try:
        image = driver.find_element(By.XPATH, "//article//img").get_attribute("src")
    except:
        image = ""

    # --- SIMPAN GAMBAR ---
    filename = f"img_{idx}.jpg"
    filepath = os.path.join(folder, filename)

    if image:
        try:
            wget.download(image, filepath)
        except:
            pass

    # --- SIMPAN DATA ---
    data.append([filename, caption, like, date, location])

    print(f"Scraped {idx+1}")

    time.sleep(3)
```

```python
with open(f"{hastag}.csv", "w", newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(["file_name", "caption", "like", "date", "location"])
    writer.writerows(data)

print("DONE ✅")
```

```python
data = []
folder = "DatasetFoto2"
os.makedirs(folder, exist_ok=True)  # WAJIB

for idx, link in enumerate(post_links):
    driver.get(link)
    time.sleep(5)

    # --- CAPTION ---
    try:
        caption = driver.find_element(By.XPATH, "//article//h1").text
    except:
        caption = ""

    # --- TANGGAL ---
    try:
        date = driver.find_element(By.TAG_NAME, "time").get_attribute("datetime")
    except:
        date = ""

    # --- LIKE (lebih akurat) ---
    try:
        like = driver.find_element(By.XPATH, "//section//span[contains(text(),'like')]").text
    except:
        try:
            like = driver.find_element(By.XPATH, "//section//span").text
        except:
            like = ""

    # --- LOKASI ---
    try:
        location = driver.find_element(By.XPATH, "//a[contains(@href,'/explore/locations/')]").text
    except:
        location = ""

    try:
        image = driver.find_element(By.XPATH, "//article//img[1]").get_attribute("src")
    except:
        image = ""

    # --- SIMPAN GAMBAR ---
    filename = f"img_{idx}.jpg"
    filepath = os.path.join(folder, filename)

    if image:
        try:
            wget.download(image, filepath)
        except:
            pass

    # --- SIMPAN DATA ---
    data.append([filename, caption, like, date, location])

    print(f"Scraped {idx+1}")

    time.sleep(3)

# --- SIMPAN CSV ---
with open(f"{hastag}.csv", "w", newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(["file_name", "caption", "like", "date", "location"])
    writer.writerows(data)

print("DONE ✅")
```

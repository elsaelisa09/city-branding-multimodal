```python
import instaloader
import pandas as pd
import os
import time
import random

# load data hasil filter
df = pd.read_excel("foto_2021_up4.xlsx")
df["File"] = df["File"].astype("object")
folder = "foto_2021_up4"
os.makedirs(folder, exist_ok=True)

# init instaloader
L = instaloader.Instaloader(
    download_videos=False,
    download_comments=False,
    save_metadata=False,
    compress_json=False
)

# pakai cookie (session kamu)
L.context._session.cookies.update({
    "sessionid": "77657383595%3AVeqtLGRWZh5dnN%3A25%3AAYhlPnALrCEYK1QuokEqnRIhtoforkBFezw_c5dEpA",
    "csrftoken": "vHgFGLNxtevk1lmOwKCEYQYmvmBjkNFD",
    "ds_user_id": "77657383595"
})
L.context.username = "eca_aapng"

# biar tidak spam error
L.context.raise_all_errors = False
counter = 1
# loop
for i, row in df.iterrows():
    try:
        url = row["Post URL"]

        if pd.isna(url) or "/p/" not in url:
            continue

        shortcode = url.split("/")[-2]
        post = instaloader.Post.from_shortcode(L.context, shortcode)

        # ======================
        # METADATA
        # ======================
        df.at[i, "Caption"] = post.caption if post.caption else ""
        df.at[i, "Likes Count"] = post.likes if post.likes else None
        try:
            location = post.location.name if post.location else None
        except:
            location = None

        df.at[i, "Location"] = location
        # ======================
        # AMBIL 1 GAMBAR (AMAN)
        # ======================
        if post.typename == "GraphSidecar":
            for node in post.get_sidecar_nodes():
                if not node.is_video:
                    url_image = node.display_url
                    break
        else:
            url_image = post.url

        # ambil extension asli
        ext = url_image.split("?")[0].split(".")[-1]
        filename = f"{str(counter).zfill(3)}.{ext}"
        counter += 1
        filepath = os.path.join(folder, filename)

        # skip kalau sudah ada
        if os.path.exists(filepath):
            df.at[i, "File"] = filename
            print(f"Skip: {filename}")
            continue

        # download
        L.download_pic(
            filename=filepath.replace(f".{ext}", ""),
            url=url_image,
            mtime=post.date
        )

        # update kolom file
        df.at[i, "File"] = filename

        print(f"OK: {filename}")

        time.sleep(random.uniform(5,10))

    except Exception as e:
        print(f"Gagal baris {i}: {e}")
        time.sleep(random.uniform(5,8))

# simpan
df.to_excel("dataset_final_up4.xlsx", index=False)
```

```python
from PIL import Image
import os
import pandas as pd

# folder gambar
folder = "ELSA"

# load excel
df = pd.read_excel("dataset_final.xlsx")

# loop semua file di folder
for file in os.listdir(folder):
    if file.lower().endswith(".webp"):
        old_path = os.path.join(folder, file)
        
        # nama baru (.jpg)
        new_file = file.replace(".webp", ".jpg")
        new_path = os.path.join(folder, new_file)

        try:
            # convert ke jpg
            img = Image.open(old_path).convert("RGB")
            img.save(new_path, "JPEG")

            # hapus file lama
            os.remove(old_path)

            print(f"Converted: {file} → {new_file}")

            # update di dataframe
            df.loc[df["File"] == file, "File"] = new_file

        except Exception as e:
            print(f"Gagal convert {file}: {e}")

# simpan kembali
df.to_excel("ALL.xlsx", index=False)
```
